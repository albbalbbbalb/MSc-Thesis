```python
np.random.seed(42)

num = 5

R = 1/3
b = 10.0**np.arange(-1,-1-num,-1,dtype=int)
n = 3000*np.ones(num, dtype=int)
xys = [(50, 70), (550,-250), (2000,550), (2900,2000), (2500,2100)]
xytexts = [(-100, 300), (850, -250), (2000, 300), (3400, 1750), (2000, 2600)]

X = np.random.random(2)-1/2
X = R*X/np.linalg.norm(X)
V = np.random.random(2)
V = V/np.linalg.norm(V)
V = V if np.dot(V,X) > 0 else -V
X = np.block([X+1/2,V])
print(X)


fig, ax = plt.subplots(1,1,figsize=(15,15))

plt.rc('xtick', labelsize=30) 
plt.rc('ytick', labelsize=30)
ax.set_aspect(1)


for idx, (i,j) in enumerate(zip(b, n)):
    XOut, XIn, xydir, dirs, entered = mag.orbit(X, R, i, int(j), maxIt=10000)
    print(entered)

    shifts = np.array([ np.sum([mag.from_where[i] for i in xy],axis=0) for xy in xydir])

    shifts = np.add.accumulate(shifts,0)

    XIn[:,:2] += shifts
    XOut[:,:2] += np.vstack([np.zeros(2),shifts])

    Xs = np.zeros((2*j-1,4))
    Xs[::2] = XOut
    Xs[1::2] = XIn

    ax.plot(Xs[:,0], Xs[:,1], "b")
    ax.annotate('b='+str(i), xy=xys[idx], xytext=xytexts[idx],
        arrowprops=dict(facecolor='black', shrink=0.05), fontsize=25)#, usetex=True)

ax.plot(Xs[0,0],Xs[0,1],"ks")
```
